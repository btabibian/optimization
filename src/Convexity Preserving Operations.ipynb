###Intersection:
If $S_1$, $S_2$ are convex then:

$S_1\cap S_2$ is convex.

Example:

We can see in the context of  $S^n_+$ :

$S^n_+=\bigcap_{x\neq0}\{X \in S^n |‌ z^TXz\geq 0\}$

for every $z\neq0$,$z^TXz$ is a (not identically zero) linear function of $X$, so the sets:

$\{X \in S^n| z^TXz\geq 0 \}$

are halfspaces in $S^n$. Thus, the positive semidifinite cone is the intesection of an infinite number of halfspaces and so convex.

> Every closed convex set S is a (usually infinite) intersection of halfspaces. In fact, a closed convex set $ is the intersection of all halfspaces that contain it:

> $S=\bigcap\{H|H \text{halfspace}, S\subseteq H\}$

###Afine functions:

$f: R^n \rightarrow R^m$ is *affine* if $f(x)=Ax+b$, where $A \in R^{m \times n}$ and $b\in R^m$, (sum of a linear function and a constant)

+ suppose $S\subseteq R^n$ is convex and $f$ and affine function, Then the image of $S$ under $f$ is convex.
+ If $f: R^k \rightarrow R^n$ is an affine function, the inverse image of $s$ under $f$:

  $f^{-1}(S) = \{x|f(x)\in S\}$
  
  is convex.
####Small Examples:

+ *Scaling* and *translations* are convex preserving:

  if $S\subseteq R^n$ is convex, $\alpha \in R$ an $a\in R^n$, then the sets $\alpha S$ and $S+a$ are convex:
  
  $\alpha S=\{\alpha x|x\in S\}$, $S+a=\{x+a|x \in S\}$
  
+ *Projection* of a convex set onto some of its  coordinates is convex.:

  if $S\subseteq R^m \times R^n$ is convex, then
  
  $T=\{x_1 \in R^m|(x_1,x_2) \in S \text{ for some } x_2 \in R^n\}$
  
  is convex.
  
+ The sum of two sets is defined as:

  $S_1+S_2=\{x+y|x\in S_1,s_2\in S_2\}$
  
  If $S_1$ and $S_2$ are convex, then $S_1+S_2$ are convex. 
  
  >To see why we note that:
  
  >if $S_1$ and $S_2$ are convex so is Cartesian product:  
  >$S_1\times S_2=\{(x_1,x_2)|x_1\in S_1, x_2 \in S_2\}$ 
  
  >The image of this set under the linear funciton $f(x_1,x_2)=x_1+x_2$ is the sum $S_1+S_2$.

###Linear Fractionals and perspective functions:

The perspective function:

\begin{align}
P: R^{n+1}\rightarrow r^n\\
\textbf{dom }P=R^n\times R_{++}&,R_{++}=\{x\in R|x\geq0\}\\
P(z,t)=z/t\\
\end{align}

This function scales or normalizes vectors by last component and then drops normalizer.

If $C \subseteq \textbf{dom} P$ is convex, then its image:

$P(C)=\{P(x)|x\in C\}

is convex.

Suppose $x = (\tilde{x},x_{n+1}),y = (\tilde{y},y_{n+1}) \in R^n+1$ and $x_{n+1}>0, y_{n+1}> 0$, then for $0 \leq \theta \leq 1$:

\begin{align}
P(\theta x+(1-\theta)y)&=\frac{\theta \tilde{x}+(1-\theta)\tilde{y}}{\theta x_{n+1}+(1-\theta)y_{n+1}}\\
&=\frac{\theta x_{n+1}}{\theta x_{n+1}+(1-\theta)y_{n+1}}\frac{\tilde{x}}{x_{n+1}}+\frac{(1-\theta) y_{n+1}}{\theta x_{n+1}+(1-\theta)y_{n+1}}\frac{\tilde{y}}{y_{n+1}}\\
&=\mu P(x)+(1-\mu)P(y)\\
\end{align}
where $\mu=\frac{\theta x_{n+1}}{\theta x_{n+1}+(1-\theta) y_{n+1}} \in [0,1]$ and monotonic.

This establishes the convexity preserving of $P$. If $C$ is convex with $C \subseteq \textbf{dom} P$ from above we have that the line segment $[P(x),P(y)]$ is in $P(C)$.
