Snow removal is an important, yet costly process in many northern cities.  Snow can  be cleared from streets and sidewalks by plowing and shoveling, but in prolonged sub-freezing temperatures, the resulting banks of accumulated snow can impede pedestrian and vehicular traffic and must be cleared.

<img src="files/imgs/montreal.png">

Large cities are divided into sectors and then snow removal operations are carried out concurrently in each sector.  

In Montreal, snow is collected and loaded onto trucks and disposed of in areas like rivers, quarries, sewer chutes, and surface holding areas.  

Different disposal sites can accommodate different amounts of snow due to physcial size or environmental restrictions on amount of snow (by salt and de-icing chemicals) that can be dumped into rivers.

```python
cap=np.array([350,250,500,400,200])
pd.DataFrame(cap,map(lambda x:"Site "+str(x),xrange(5)),columns=["Capacity of Disposal Sites, in 1000s"])
```

The cost of removing and disposing of snow depends mainly on the distance it must be trucked.  The city of Montreal uses the straight-line distance between the center of each sector to each of the various disposal sites as an approximation of the cost involved in transporting snow between these locations.

Each sector may only be assigned to one disposal site.


```python
C=np.array([[3.4,1.4,4.9,7.4,9.3],
             [2.4,2.1,8.3,9.1,8.8],
             [1.4,2.9,3.7,9.4,8.6],
             [2.6,3.6,4.5,8.2,8.9],
             [1.5,3.1,2.1,7.9,8.8],
             [4.2,4.9,6.5,7.7,6.1],
             [4.8,6.2,9.9,6.2,5.7],
             [5.4,6.0,5.2,7.6,4.9],
             [3.1,4.1,6.6,7.5,7.2],
             [3.2,6.5,7.1,6.0,8.3]])
pd.DataFrame(C,index=map(lambda x:"Section "+str(x),xrange(10)),columns=map(lambda x:"Site "+str(x),xrange(5)))
```

Using past snowfall data, the city of Montreal estimated the annual volume of snow requiring removal in each sector as four times the length of streets in the sectors in meters (it is assumed that each linear meter of street generates four cubic meters of snow to remove over an entire year)

```python
b=np.array([153,152,154,138,127,129,111,110,130,135])
pd.DataFrame(b,map(lambda x:"Section "+str(x),xrange(10)),columns=["Estimated Annual Snow Removal Requirements in 1000s"])
```

### Linear Programming

The first step is to compute the solution for continues variables using CVX Library. This library is easy to use, however does not support Integer Linear Programming. We will use another library for Integer Linear Programming.

```python
def construct_unit(i):
    unit=np.zeros((15,5))
    unit[10,0]=-1
    unit[11,1]=-1
    unit[12,2]=-1
    unit[13,3]=-1
    unit[14,4]=-1
    unit[i,:]=1
    return unit
A=np.concatenate(map(construct_unit,xrange(10)),axis=1)
x_1=cvp.Variable(rows=50)
constraints = [ x_1>=0,(A*x_1)[:10,:] == b.reshape(10,1),-(A*x_1)[10:,:]<=cap.reshape(5) ]
objective1 = cvp.Minimize(C.reshape(50,1).T*x_1)
p1 = cvp.Problem(objective1, constraints)
res=p1.solve()
print "Status:" + str(res)
vals= np.array(x_1.value)
links= np.arange(50)[vals.reshape(50)>0.01]

for i in zip(links,vals[vals.reshape(50)>0.01]):
    print "x"+str(int(i[0])/5)+"_"+str(int(i[1])%5)+"= %1.4f" %(i[1][0]/b[int(i[0])/5])
print "Total Cost = ", objective1.value
```

We can use CVX to compute Lagrange multipliers and rule out those constraints(sites) which do not reduce costs if their capacities are increased.

The effect of change on constraints and its relation to dual variables are discussed in Convex Optimization book by Stephen Boyd under chapter "Dual Variables". Here we discuss it briefly:

>#### Optimization problem
>$\begin{align}
\text{minimize }&f_0(x)\\
\text{subject to }&f_i(x)\leq 0\\
&h_i(x)=0\\
\end{align}$

>##### Corresponding dual:
$\begin{align}
\text{maximize }&g(\lambda,\nu)\\
\lambda\succeq 0
\end{align}$


>#### perturbed Optimization problem
>$\begin{align}
\text{minimize }&f_0(x)\\
\text{subject to }&f_i(x)\leq u_i\\
&h_i(x)=v_i\\
\end{align}$

>##### Corresponding dual:
>$\begin{align}
\text{maximize }&g(\lambda,\nu)-u^T\lambda-v^T\nu \\
\lambda\succeq 0
\end{align}$

> Assume strong duality holds for for unperturbed problem, $\lambda^*$,$\nu^*$ are dual optimal for unpurturbed problem.
> + if $\lambda^*$ large: $p^*$ increases greatly if we tighten constraint i $u_i<0$
> + if $\lambda^*$ small: $p^*$ does not decrease much if we loosen constraint i $u_i<0$

Above considerations help us rule out last three constraints since they are very small compared to first two.

### Integer Linear Programming

Finally we solve the integer version of the problem, here we use a "PuLP" which is an interface to GLPK library.

```python
import pulp as plp
from operator import add
prob = plp.LpProblem("Problem",plp.LpMinimize)

C=np.array([[3.4,1.4,4.9,7.4,9.3],
             [2.4,2.1,8.3,9.1,8.8],
             [1.4,2.9,3.7,9.4,8.6],
             [2.6,3.6,4.5,8.2,8.9],
             [1.5,3.1,2.1,7.9,8.8],
             [4.2,4.9,6.5,7.7,6.1],
             [4.8,6.2,9.9,6.2,5.7],
             [5.4,6.0,5.2,7.6,4.9],
             [3.1,4.1,6.6,7.5,7.2],
             [3.2,6.5,7.1,6.0,8.3]])
b=np.array([153,152,154,138,127,129,111,110,130,135])
cap=[350,250,500,400,200]
var=[]
val=None
const=None
caps=[[],[],[],[],[],[]]
for i in xrange(10):
    current=[]
    for j in xrange(5):
        x=plp.LpVariable("x"+str(i)+"_"+str(j),0,1,plp.LpInteger)
        var+=[x]
        current+=[x]
        if val is None:
            val=C[i,j]*b[i]*x
        else:
            val+=C[i,j]*b[i]*x
        caps[j]+=[x]
    prob+=reduce(add, current)==1
for i in xrange(5):
    con=None
    for j in zip(caps[i],b):
        if con is None:
            con=j[0]*j[1]
        else:
            con+=j[0]*j[1]
    prob+=con<=cap[i]
prob+=val
#prob+=const
prob.solve()
print "Status:", plp.LpStatus[prob.status]
for v in prob.variables():
    if v.varValue>0:
        print v.name, "=", v.varValue
print "Total Cost = ", plp.value(prob.objective)
```

### Sensitivity Analysis
Suppose Montreal can increase the capacity of a single disposal site by 100,000 cubic meters.  Which disposal site’s capacity (if any) should be increased and how much should the city be willing to pay to obtain this extra disposal capacity?

The solution below demonstrates that we are better off increasing the capacity on site1, variable $y$ shows the index of the site, which matches our prediction from earlier section.

```python
import pulp as plp
from operator import add
prob = plp.LpProblem("Problem",plp.LpMinimize)

C=np.array([[3.4,1.4,4.9,7.4,9.3],
             [2.4,2.1,8.3,9.1,8.8],
             [1.4,2.9,3.7,9.4,8.6],
             [2.6,3.6,4.5,8.2,8.9],
             [1.5,3.1,2.1,7.9,8.8],
             [4.2,4.9,6.5,7.7,6.1],
             [4.8,6.2,9.9,6.2,5.7],
             [5.4,6.0,5.2,7.6,4.9],
             [3.1,4.1,6.6,7.5,7.2],
             [3.2,6.5,7.1,6.0,8.3]])
b=np.array([153,152,154,138,127,129,111,110,130,135])
cap=[350,250,500,400,200]
var=[]
val=None
const=None
caps=[[],[],[],[],[],[]]
for i in xrange(10):
    current=[]
    for j in xrange(5):
        x=plp.LpVariable("x"+str(i)+"_"+str(j),0,1,plp.LpInteger)
        var+=[x]
        current+=[x]
        if val is None:
            val=C[i,j]*b[i]*x
        else:
            val+=C[i,j]*b[i]*x
        caps[j]+=[x]
    prob+=reduce(add, current)==1
lst_y=[]
for i in xrange(5):
    con=None
    y=plp.LpVariable("y"+str(i),0,1,plp.LpInteger)
    for j in zip(caps[i],b):
        if con is None:
            con=j[0]*j[1]
        else:
            con+=j[0]*j[1]
    prob+=con<=cap[i]+y*100
    lst_y+=[y]
prob+=reduce(add, lst_y)==1
prob+=val
#prob+=const
prob.solve()
print "Status:", plp.LpStatus[prob.status]
for v in prob.variables():
    if v.varValue>0:
        print v.name, "=", v.varValue
print "Total Cost = ", plp.value(prob.objective)
```
